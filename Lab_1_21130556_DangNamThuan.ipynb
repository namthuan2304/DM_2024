#1.1. Implement Python functions as follows:
1.   S(n)=1 - 2 + 3 - 4+ … + ((-1) ^ (n + 1)).n , 		n>0
2.   S(n)=1 ^ 2 + 2^2 + 3^2 + ... + n^2 , 			n>0


```python
#1. S(n)=1 - 2 + 3 - 4+ … + ((-1) ^ (n + 1)).n , n>0
def S1():
    n = int(input("Enter a positive integer value for n: "))
    if n <= 0:
        print("n must be a positive integer.")
        return

    re = 0
    for i in range(1, n+1):
        re += ((-1) ** (i + 1)) * i
    print(f"S1({n}) = {re}")

#2. S(n)=1 ^ 2 + 2^2 + 3^2 + ... + n^2 , n>0
def S2():
    n = int(input("Enter a positive integer value for n: "))
    if n <= 0:
        print("n must be a positive integer.")
        return

    re = 0
    for i in range(1, n+1):
        re += i**2
    print(f"S2({n}) = {re}")

S1()
S2()
```

#1.2. For a given class diagram (in Java) as follows:

Implements the following methods:


1.   Count the number of employees having birthYear equals to a given year
2.   Find the employee with highest age
3.   Summary employees by birth years. Report results as a dictionary structure.

```python
from datetime import datetime
class Employee:
      def __init__(self, id, name, birthYear, salaryRate):
        self.__id = id
        self.__name = name
        self.__birthYear = birthYear
        self.__salaryRate = salaryRate

      def get_birthYear(self):
        return self.__birthYear

      def get_ages(self):
        current_year = datetime.now().year
        return current_year - self.__birthYear

      def __repr__(self):
        return self.__name
```

```python
class Department:
      def __init__(self, name):
        self.__name = name
        self.__employees = list()

      def set_employees(self, employees):
        if employees:
            self.__employees.extend(employees)

      #1. Count the number of employees having birthYear equals to a given year
      def countBirthYearEqual(self, year):
        if not self.__employees:
          return None
        re = 0
        for employee in self.__employees:
          if employee.get_birthYear() == year:
            re += 1
        return re

      #2. Find the employee with highest age
      def findHighestAge(self):
        if not self.__employees:
          return None
        max_age = -1
        employees_max_age = None
        for employee in self.__employees:
          if employee.get_ages() > max_age:
            max_age = employee.get_ages()
            employees_max_age = employee

        return employees_max_age

      #3. Summary employees by birth years. Report results as a dictionary structure.
      def filterByBirthYear(self):
        if not self.__employees:
          return None
        s = set()
        re = dict()
        for employee in self.__employees:
          s.add(employee.get_birthYear())
        for year in s:
          re[year] = self.countBirthYearEqual(year)
        return re
```

```python
e1 = Employee(1, "John", 1994, 5.5)
e2 = Employee(2, "Tom", 1996, 5.6)
e3 = Employee(3, "Iris", 1992, 5.8)
e4 = Employee(4, "Elva", 1992, 5.9)
e5 = Employee(5, "Duke", 1995, 6.3)

l = list()
l.append(e1)
l.append(e2)
l.append(e3)
l.append(e4)
l.append(e5)

department = Department("Noi vu")
department.set_employees(l)

print(department.countBirthYearEqual(1992))
print(department.findHighestAge())
department.filterByBirthYear()
```
